```python
# calculate the mean of Sepal Length
sepallength_mean= df['sepal_length'].mean()
print('The mean of Sepal Length', sepallength_mean)
```

```python
# calculate the mean of Sepal Width
sepalwidth_mean= df['sepal_width'].mean()
print('The mean of Sepal Width', sepalwidth_mean)
```

```python
# calculate the mean of Petal Length
petallength_mean= df['petal_length'].mean()
print('The mean of Petal Length', petallength_mean)
```

```python
# calculate the mean of Petal Width
petalwidth_mean= df['petal_width'].mean()
print('The mean of Petal Width', petalwidth_mean)
```

## Replacing NAN with the use of mean

```python
import numpy as np
df['sepal_length'].replace(np.nan,sepallength_mean,inplace=True)
```

```python
df.head(5)
```

```python
df['sepal_width'].replace(np.nan,sepalwidth_mean,inplace=True)
```

```python
df.head(5)
```

```python
df['petal_length'].replace(np.nan,petallength_mean,inplace=True)
```

```python
df.head(5)
```

```python
X=df.iloc[:,0]
X
```

```python
Y=df.iloc[:,1]
Y
```

## Training & Testing for Iris DataSet

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
print('X Train:')
x_train = np.array(x_train).reshape(-1,1)
print(x_train)
```

```python
print('X Test:')
x_test = np.array(x_test).reshape(-1,1)
print(x_test)
```

```python
print('Y Train:')
y_train = np.array(y_train).reshape(-1,1)
print(y_train)
```

```python
print('Y Test:')
y_test = np.array(y_test).reshape(-1,1)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_test=lr.predict(x_test)
y_pred_test
```
